```python
import numpy as np
import tensorflow as tf
from tensorflow.keras.datasets import imdb
from tensorflow.keras.preprocessing import sequence
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Embedding,SimpleRNN,Dense
```

```python
max_features=10000## vocab size
(X_train,y_train),(X_test,y_test)=imdb.load_data(num_words=max_features)
print(X_train.shape)
```

```python
X_train[0],y_train[0]
```

```python
#inspect
sample_review=X_train[0]
sample_label=y_train[0]
print(f"sample review as int :{sample_review}")
print(f"saple label :{sample_label}")
```

```python
from tensorflow.keras.preprocessing import sequence
max_len=500

X_train=sequence.pad_sequences(X_train,maxlen=max_len)
X_test=sequence.pad_sequences(X_test,maxlen=max_len)
X_train
```

```python
## train our simple rnn
model=Sequential()
model.add(Embedding(max_features,128,input_length=max_len)) ### embedding layer
model.add(SimpleRNN(128,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
## create an instance of  early stopping
from tensorflow.keras.callbacks import EarlyStopping
earlystopping=EarlyStopping(monitor='val_loss',patience=5,restore_best_weights=True)
earlystopping
```

```python
history=model.fit(
    X_train,y_train,epochs=10,batch_size=32,validation_split=0.2,
    callbacks=[earlystopping]
)
```

```python
##save model file
model.save('simple_rnn_imbd.h5')
```
